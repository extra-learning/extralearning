```python
# Both multi-class and binary classification are containg 
# within extralearning.supervised.Classification class

from extralearning.supervised import Classification
from sklearn.datasets import make_classification

# Constant random state
RANDOM_STATE = 102
```

```python
# Creating classification data

X, y = make_classification(n_samples = 1000,
                           n_features = 4,
                           n_informative = 2,
                           n_redundant = 0,
                           random_state = RANDOM_STATE,
                           shuffle = False)
```

```python
# Creating the extralearning binary classifier

model = Classification(random_state = RANDOM_STATE,
                       n_jobs = -1,
                       ignore_warnings = True)
```

----
Once you've created the classifier there is two types of methods that are available to use, the ones that are provided from the parent EstimatorClass class and the ones that are class specific:

### EstimatorClass Methods:
Those that come from the parent class are available for all the extralearning.supervised classes (Classification/Regression).

- **get_estimators():** Return a list of tuples with (name, estimator) format of all estimator available for the class.
- **add_estimator(estimator: tuple):** Adds a new estimator (name, estimator) to the list of estimators.
- **remove_estimator(estimator: (str or int)):** Removes an estimator from the list, str if want to remove by name or int in case of index position removal.
- **get_params():** Get parameters for each estimator on the list.
- **pass_params(estimator: tuple, overwrite: bool):** Set the parameters of one estimator of the list.

### Classification Methods:

- **reset_estimators():** Resets the estimators to the original list.
- **fit_train(X, y, CV, CV_Stratified, CV_params, verbose):** Fits then trains all the estimators with the provided X, y and captures metrics.
- **summary(pandas):** Returns the summary of the training.

```python
# Fit and train the classifier

model.fit_train(X, y, CV = 2, CV_Stratified = False, verbose = True)
```

```python
# Creating summary

model.summary()
```
